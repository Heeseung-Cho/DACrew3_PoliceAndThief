### Optimizer를 사용해서 loss function의 최저점을 찾는 과정

1. 경사하강법으로 방향과 보폭을 정한다. (화살표의 방향과 길이라고 생각하면 된다)
2. 방향과 보폭을 정하고 내려가면 backpropagation(역전파)를 한다. 
#### backpropagation이란?
- 경사하강법으로 내려가던 화살표를 다시 왔던 길로 올라가면서 weight를 update시켜주는 것이다.
- 중요도를 생각해서 가중치를 더 주는 과정.
