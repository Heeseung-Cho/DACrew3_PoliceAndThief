## 1. Numpy
- NumpyÎäî ÏàòÌïôÍ≥º Í≥ºÌïô Ïó∞ÏÇ∞ÏùÑ ÏúÑÌïú ÌååÏù¥Ïç¨ Ìå®ÌÇ§ÏßÄÎ•º ÎßêÌï©ÎãàÎã§.
- NumpyÎäî Í∏∞Î≥∏ Ïó∞ÏÇ∞ Î∞è Î≤°ÌÑ∞, ÌñâÎ†¨ Îì± Îã§ÏñëÌïú Ïó∞ÏÇ∞Ïù¥ ÏÜêÏâΩÍ≤å Í∞ÄÎä•Ìï©ÎãàÎã§.

```python
import numpy as np 
```

ÎÑòÌååÏù¥Î•º ÌÜµÌï¥ Îã§ÏñëÌïú Ïó∞ÏÇ∞ ÏàòÌñâÏù¥ Í∞ÄÎä•ÌïòÎã§Í≥† ÌñàÏäµÎãàÎã§. ÌïòÏßÄÎßå Í∏∞Ï°¥Ïùò Î¶¨Ïä§Ìä∏ ÌòïÌÉúÎ°úÎäî Í∏∞Î≥∏Ï†ÅÏù∏ Ïó∞ÏÇ∞Ïù¥ Î∂àÍ∞ÄÎä•Ìï©ÎãàÎã§. Ïó∞ÏÇ∞ÏùÑ ÏúÑÌï¥ Î®ºÏ†Ä ÌååÏù¥Ïç¨ÏóêÏÑú Î¶¨Ïä§Ìä∏Î•º NumpyÎ•º ÌÜµÌï¥ Î∞∞Ïó¥ ÌòïÌÉúÎ°ú Î≥ÄÌôòÌïòÎäî Î∞©Î≤ïÏùÑ ÏïåÏïÑÎ¥ÖÏãúÎã§.

np.arrayÎ•º Ïù¥Ïö©ÌïòÎ©¥ Î∞∞Ïó¥ ÌòïÌÉúÎ°ú Î≥ÄÌôòÌï† Ïàò ÏûàÏäµÎãàÎã§.

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
```

ÏïûÏÑú ÏÉùÏÑ±Ìïú Î∞∞Ïó¥ÏùÑ Ïù¥Ïö©Ìï¥ Ïó∞ÏÇ∞Ïù¥ Í∞ÄÎä•Ìï©ÎãàÎã§. ÎÑòÌååÏù¥ Î∞∞Ïó¥ÏóêÏÑú Ïó∞ÏÇ∞ÏùÄ elementwise(ÏöîÏÜåÎ≤å)Ïó∞ÏÇ∞ÏûÖÎãàÎã§. Ï¶â, Í∞Å Î∞∞Ïó¥Ïóê ÎåÄÏùëÎêòÎäî ÏõêÏÜåÎì§ÎÅºÎ¶¨ Ïó∞ÏÇ∞Ïù¥ ÏßÑÌñâÎê©ÎãàÎã§. Ïù¥Îûò ÏòàÏãúÎ•º Î≥¥Í≤†ÏäµÎãàÎã§.

```python
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

ÏïûÏÑú Ï†ïÏùòÌïú Î∞∞Ïó¥Îì§Ïóê ÏÇ¨ÏπôÏó∞ÏÇ∞ÏùÑ ÏàòÌñâÌñàÏäµÎãàÎã§. Î∞∞Ïó¥Ïùò Í∞Å ÏúÑÏπòÏóê ÎåÄÏùëÎêòÎäî ÏõêÏÜåÎì§ÎÅºÎ¶¨ Í≥ÑÏÇ∞Îêú Í≤ÉÏùÑ Ïïå Ïàò ÏûàÏäµÎãàÎã§.

  - ÎÇ¥Ï†Å : ÎÇ¥Ï†ÅÏùò Ï†ÅÏùÄ Í≥±ÌïúÎã§ÎùºÎäî ÏùòÎØ∏Î°ú Î≤°ÌÑ∞ ÏûêÏ≤¥Î•º ÏàòÏ≤òÎüº Í≥±Ìïú Í∞íÏùÑ ÎÇòÌÉÄÎÇ∏Îã§Í≥† Î≥º Ïàò ÏûàÏäµÎãàÎã§.

ÏúÑ Ï≤´Î≤àÏß∏ Ïù¥ÎØ∏ÏßÄÍ∞Ä 2Ï∞®Ïõê Î∞∞Ïó¥ÏùÑ Î≥¥Í∏∞ ÏâΩÍ≤å ÎÇòÌÉÄÎÇ∏ Í≤ÉÏûÖÎãàÎã§. ÌïòÏßÄÎßå Ïã§Ï†ú Ïª¥Ìì®ÌÑ∞Ïùò Î©îÎ™®Î¶¨Îäî ÏúÑÏùò Í∞ÑÎã®Ìïú Í∑∏Î¶ºÍ≥º Îã¨Î¶¨ ÏûÖÏ≤¥Ï†Å Í≥µÍ∞ÑÏù¥ ÏïÑÎãå ÏÑ†Ìòï Í≥µÍ∞ÑÏù¥Í∏∞ ÎïåÎ¨∏Ïóê Ïã§Ï†úÎ°úÎäî ÏïÑÎûò ÏÇ¨ÏßÑÍ≥º Í∞ôÏù¥ Ï†ÄÏû•Îê©ÎãàÎã§.

Í∑∏Îüº Ïù¥Î≤àÏóêÎäî ÏúÑÏùò ÏÇ¨ÏßÑÏ≤òÎüº 2Ï∞®Ïõê Î∞∞Ïó¥ÏùÑ ÏÉùÏÑ±Ìï¥Î≥¥ÎèÑÎ°ù ÌïòÍ≤†ÏäµÎãàÎã§. 2Ï∞®Ïõê Î∞∞Ïó¥ÎèÑ 1Ï∞®Ïõê Î∞∞Ïó¥Í≥º ÎßàÏ∞¨Í∞ÄÏßÄÎ°ú elementwise(ÏöîÏÜåÎ≥Ñ)Ïó∞ÏÇ∞Ïù¥ ÏßÑÌñâÎê©ÎãàÎã§.

```python
mat1 = np.array([[1,2],[3,4]])
mat2 = np.array([[5,6],[7,8]])
```

```python
size = 1000
a = np.random.rand(size, size) # size * sizeÏùò ÌñâÎ†¨ ÏÉùÏÑ±.
b = np.random.rand(size, size)

start = time.time()

x = np.dot(a,b) # Îëê 600*600 ÌñâÎ†¨Ïùò ÌñâÎ†¨Í≥± Í≥ÑÏÇ∞Ïóê ÏïΩ 1.340s ÏÜåÏöî.
print("ÎÑòÌååÏù¥ Ïó∞ÏÇ∞ ÏÜçÎèÑ:",time.time() - start)
```

```python
aa = torch.FloatTensor(a)
bb = torch.FloatTensor(b)

start = time.time()

cc = torch.matmul(aa, bb)
print("ÌååÏù¥ÌÜ†Ïπò Ïó∞ÏÇ∞ ÏÜçÎèÑ(CPU):",time.time() - start)
```

```python
aaa = torch.cuda.FloatTensor(a)
bbb = torch.cuda.FloatTensor(b)

start = time.time()

ccc = torch.matmul(aaa, bbb)
print("ÌÜ†Ïπò Ïó∞ÏÇ∞ ÏÜçÎèÑ(GPU):",time.time() - start) # Îß•Î∂Å Ïó∞ÏÇ∞Ïùò ÌïúÍ≥ÑÎ°ú ÏßÑÌñâ x
```

### Tensor Î≥ÄÌôò

TensorÎäî Ìòï Î≥ÄÌôòÏù¥ ÏûêÏú†Î°≠ÏäµÎãàÎã§. Ï¶â, ÏÉàÎ°úÏö¥ Ï∞®ÏõêÏùÑ Ï∂îÍ∞ÄÌïòÍ±∞ÎÇò Í∏∞Ï°¥Ïùò Ï∞®ÏõêÏùÑ Ï∂ïÏÜåÌïòÎäîÎì±Ïùò ÌòïÎ≥ÄÌôòÏù¥ ÏûêÏú†Î°≠ÏäµÎãàÎã§.

Ïù¥Ïô∏ÏóêÎèÑ Í∞ÅÍ∞ÅÏùò ÌÖêÏÑúÎ•º Í≤∞Ìï©ÌïòÍ±∞ÎÇò ÎëêÍ∞úÎ°ú Î∂ÑÎ¶¨ÌïòÎäîÍ≤É ÎòêÌïú Í∞ÄÎä•Ìï©ÎãàÎã§. 

```python
x1 = torch.FloatTensor(1,32,32)
x2 = torch.squeeze(x1)  # torch.Size[(32, 32)]

x1.size(),x2.size()
```

```python
x1 = torch.FloatTensor(32,32)
x2 = torch.unsqueeze(x1, dim=0)
print(x1.size(),x2.size()) # torch.Size([32, 32]) torch.Size([1, 32, 32])
x3 = torch.unsqueeze(x1, dim=1)
print(x1.size(),x3.size()) # torch.Size([32, 32]) torch.Size([32, 1, 32])
```

```python
x = torch.FloatTensor([[1,2,3],[4,5,6]])
y = torch.FloatTensor([[-1,-2,-3],[-4,-5,-6]])
xy1 = torch.cat([x,y], dim = 0)  # 0Î≤àÏß∏Ïóê Í≤∞Ìï©
xy2 = torch.cat([x,y], dim = 1)  # 1Î≤àÏß∏Ïóê Í≤∞Ìï©
print(x.size(), y.size(), xy1.size(), xy2.size())
```

```python
x1,x2 = torch.split(xy1,2,dim=0)
print(x1)
print(x2)
```

```python
x1,x2 = torch.split(xy1,2,dim=1)
print(x1)
print(x2)
```

### Pytorch Ïù¥ÎØ∏ÏßÄ Î∂àÎü¨Ïò§Í∏∞

ÎßàÏßÄÎßâÏúºÎ°ú torchÏóêÏÑú Î™®Îç∏ ÌïôÏäµÏùÑ ÏúÑÌïú Ïù¥ÎØ∏ÏßÄ Îç∞Ïù¥ÌÑ∞Î•º Î∂àÎü¨Ïò§Îäî Í≥ºÏ†ïÏóê ÎåÄÌï¥ ÌïôÏäµÌïòÍ≤†ÏäµÎãàÎã§.

Ïù¥Î≤àÏóê Ïù¥Ïö©Ìï† Îç∞Ïù¥ÌÑ∞ÏÖãÏùÄ Íµ¨Í∏ÄÏóêÏÑú Ï†úÍ≥µÌïòÎäî 5Ï¢ÖÏùò ÍΩÜ Ïù¥ÎØ∏ÏßÄÍ∞Ä Îã¥Í∏¥ Îç∞Ïù¥ÌÑ∞ÏÖãÏûÖÎãàÎã§. Ïù¥Î•º Î™®Îç∏Ïóê ÌïôÏäµÏù¥ Í∞ÄÎä•Ìïú Îç∞Ïù¥ÌÑ∞ÏÖãÏúºÎ°ú Ï†ÑÏ≤òÎ¶¨ ÌïòÎäî Í≥ºÏ†ïÏùÑ ÏßÑÌñâÌïòÎäî Î∞©Î≤ïÏùÑ ÏïåÏïÑÎ≥¥ÎèÑÎ°ù ÌïòÍ≤†ÏäµÎãàÎã§.

```python
import tensorflow as tf
import pathlib
dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
data_dir = tf.keras.utils.get_file(origin=dataset_url, 
                                   fname='flower_photos', 
                                   untar=True,
                                   )
data_dir = pathlib.Path(data_dir)
```

```python
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, random_split
from torchvision.transforms import transforms as T
t = T.Compose([T.Resize((224,224)),T.ToTensor()])
dataset = ImageFolder(root = data_dir,transform = t)
```

torchÏóêÏÑú Î∂àÎü¨Ïò® Îç∞Ïù¥ÌÑ∞Îäî numpyÏóêÏÑúÏùò Í≤ΩÏö∞ÏôÄ Îã§Î•∏ ÏàúÏÑúÎ•º Í∞ÄÏßëÎãàÎã§. h w c -> c h w

```python
dataset[0][0].shape ## c,h,wÎ°ú Î∞îÎÄú
```

```python
train_size = int(dataset.__len__() * 0.8)
valid_size = dataset.__len__() - train_size

training_data, valid_data = random_split(dataset, [train_size, valid_size])

train_dataloader = DataLoader(training_data, batch_size=16, shuffle=True)
valid_dataloader = DataLoader(valid_data, batch_size=16, shuffle=False)
```

Í∞ÅÍ∞ÅÏùò Îç∞Ïù¥ÌÑ∞Î•º ÌõàÎ†®Í≥º Í≤ÄÏ¶ùÏö©ÏúºÎ°ú Î∂ÑÎ•òÌï¥ Î≥ÄÏàòÏóê Îã¥ÏïÑÏ§çÎãàÎã§. Ïù¥Î•º Îç∞Ïù¥ÌÑ∞Î°úÎçîÎ•º Ïù¥Ïö©Ìï¥ ÏõêÌïòÎäî batchÌÅ¨Í∏∞ÎßåÌÅº Ï†ÅÏû¨Ìï©ÎãàÎã§.

Îç∞Ïù¥ÌÑ∞Î°úÎçîÎ•º ÌôúÏö©Ìï¥ Í∞ÅÍ∞ÅÏùò Ïù¥ÎØ∏ÏßÄÎç∞Ïù¥ÌÑ∞Î•º Î∂àÎü¨Ïò¨Îïå batch_sizeÏóêÏÑú Ï†ïÏùòÌïú ÎßåÌÅº Î∂àÎü¨Ïò¨ Ïàò ÏûàÏäµÎãàÎã§. 

ÌïúÍ∞úÏùò batch_sizeÎ•º Î∂àÎü¨ÏôÄ ÏãúÍ∞ÅÌôîÎ•º Ìï¥Î≥¥ÎèÑÎ°ù ÌïòÍ≤†ÏäµÎãàÎã§.

```python
class_names = ['daisy','dandelion','roses','sunflowers','tulibs']
plt.figure(figsize=(10, 10))
images, labels = next(iter(train_dataloader))
for i in range(16):
  ax = plt.subplot(4, 4, i + 1)
  img = images[i]
  img = (img - img.min())/(img.max() - img.min())
  plt.imshow(img.permute(1, 2, 0).numpy())
  plt.title(f'Class:{class_names[labels[i]]}')
  plt.axis("off")
```
