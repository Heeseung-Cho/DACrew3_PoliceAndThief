- 넘파이를 통해 다양한 연산 수행이 가능하다고 했습니다. 하지만 기존의 리스트 형태로는 기본적인 연산이 불가능합니다. 연산을 위해 먼저 파이썬에서 리스트를 Numpy를 통해 배열 형태로 변환하는 방법을 알아봅시다.
- np.array를 이용하면 배열 형태로 변환할 수 있습니다.

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
```

- 앞서 생성한 배열을 이용해 연산이 가능합니다. 넘파이 배열에서 연산은 elementwise(요소벌)연산입니다. 즉, 각 배열에 대응되는 원소들끼리 연산이 진행됩니다. 이래 예시를 보겠습니다.

```python
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

- 앞서 정의한 배열들에 사칙연산을 수행했습니다. 배열의 각 위치에 대응되는 원소들끼리 계산된 것을 알 수 있습니다. 
- 또한 1차원 넘파이 배열을 벡터로써의 기능도 겸합니다. 즉 내적 등의 벡터 연산이 가능합니다.
  - 내적 : 내적의 적은 곱한다라는 의미로 벡터 자체를 수처럼 곱한 값을 나타낸다고 볼 수 있습니다.

```python
np.dot(arr1,arr2) #내적
```

- 그럼 이번에는 위의 사진처럼 2차원 배열을 생성해보도록 하겠습니다. 2차원 배열도 1차원 배열과 마찬가지로 elementwise(요소별)연산이 진행됩니다.

```python
mat1 = np.array([[1,2],[3,4]])
mat2 = np.array([[5,6],[7,8]])
```

```python
print(mat1 + mat2)
print(mat1 - mat2)
print(mat1 * mat2)
print(mat1 / mat2)
```

- 2차원 넘파이 배열은 행렬의 기능또한 겸합니다. 따라서 행렬곱과 같은 행렬연산이 가능합니다.

```python
np.matmul(mat1,mat2) ##행렬의 곱셈
```

## 3. Tensor(Pytorch)

- 텐서란 다차원 배열을 일반화해서 표현하는 개념을 말합니다. 1차원의 데이터는 벡터 2차원 데이터는 행렬이라고 명칭합니다.
- 1차원 2차원인 벡터와 행렬 및 3차원 이상의 다차원 숫자로 이루어진 데이터를 **'벡터'**라고 합니다.
- 딥러닝에서 주로 사용되는 연산들은 이러한 큰 데이터에 간단한 곱하기와 더하기로 이루어져 있습니다. 하지만 계산 시행횟수가 많기 때문에 연산을 시행하는데 시간과 자원이 많이 소모됩니다. GPU 등 연산 가속장치들에 연동해 연산속도를 증가시킬 수 있습니다. 이를 위해 Tensor를 사용합니다.
- 간단한 예제를 위해 아래와 같이 사용한 모듈을 임포트 해줍니다.

```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import time
```

- 이제 넘파이와 텐서의 연산속도를 비교해보도록 하겠습니다.
- 넘파이로도 3차원 이상의 데이터를 표현하는것이 가능하지만, 텐서로 연산을 진행하면 훨씬 빠른 속도로 진행이 가능합니다.(GPU를 통한 연산속도 가속이 가능)

```python
size = 1000
a = np.random.rand(size, size) # size * size의 행렬 생성.
b = np.random.rand(size, size)

start = time.time()

x = np.dot(a,b) # 두 600*600 행렬의 행렬곱 계산에 약 1.340s 소요.
print("넘파이 연산 속도:",time.time() - start)
```

```python
aa = torch.FloatTensor(a)
bb = torch.FloatTensor(b)

start = time.time()

cc = torch.matmul(aa, bb)
print("파이토치 연산 속도(CPU):",time.time() - start)
```

```python
aaa = torch.cuda.FloatTensor(a)
bbb = torch.cuda.FloatTensor(b)

start = time.time()

ccc = torch.matmul(aaa, bbb)
print("토치 연산 속도(GPU):",time.time() - start) # 맥북 연산의 한계로 진행 x
```

- 위의 결과처럼 tensor를 이용하는 pytorch에서의 GPU를 통한 연산속도가 훨씬 빠른것을 볼 수 있었습니다.

### Tensor의 생성 및 연산

- Tensor를 새로 생성할 수도 있지만 기존의 넘파이 배열을 tensor로 변환이 가능하고, 반대로 tensor를 넘파이 배열로 변환또한 가능합니다. 
- 텐서는 넘파이 배열과 마찬가지로 연산이 가능합니다. 그리고 넘파이 배열과 마찬가지로 난수생성도 가능합니다.

```python
x = np.array([[1,2,3],[4,5,6]], dtype=int) ## numpy array
x_tensor = torch.from_numpy(x)             ## numpy to tensor
print(x_tensor, x_tensor.type())
x_numpy = x_tensor.numpy()                 ## tensor to numpy
print(x_numpy)
```

```python
x_float = torch.FloatTensor(x)
print(x_float, x_float.type())
```

```python
x = torch.rand(2,3)
x.shape
```

```python
torch.randn(2,3)
torch.zeros(2,3)
torch.ones(2,3)
```

- 앞서 살펴본 것처럼 tensor 형태로 딥러닝을 수행하게 되면 기존의 연산의 주체가 CPU인데 반해 tensor를 이용하게 되면 GPU를 통한 연산속도의 가속이 가능하게 됩니다. 이를 선택하여 수행도 가능합니다. 방법은 아래와 같습니다.

```python
cpu = torch.device('cpu')
gpu = torch.device('cuda')
x_float.to(gpu)             ## GPU에 할당
x_float.to(cpu)             ## CPU에 할당
```

```python
x = torch.FloatTensor([[1,2,3],[4,5,6]])
y = torch.FloatTensor([[-1,-2,-3],[-4,-5,-6]])
```

```python
torch.add(x,y) ## x+y
torch.mul(x,y) ## x*y
torch.div(x,y) ## x/y
torch.pow(x,2) ## x**2
```

```python
torch.mm(x,y.T) ## 행렬곱
torch.log(x)    ## log(x)
torch.exp(x)    ## exp(x)
```

- 이처럼 넘파이 배열과 같이 연산을 진행할 수 있습니다.

### Tensor 변환

- Tensor는 형 변환이 자유롭습니다. 즉, 새로운 차원을 추가하거나 기존의 차원을 축소하는등의 형변환이 자유롭습니다.
- 이외에도 각각의 텐서를 결합하거나 두개로 분리하는것 또한 가능합니다. 

```python
x1 = torch.FloatTensor(1,32,32)
x2 = torch.squeeze(x1)  # torch.Size[(32, 32)]

x1.size(),x2.size()
```

```python
x1 = torch.FloatTensor(32,32)
x2 = torch.unsqueeze(x1, dim=0)
print(x1.size(),x2.size()) # torch.Size([32, 32]) torch.Size([1, 32, 32])
x3 = torch.unsqueeze(x1, dim=1)
print(x1.size(),x3.size()) # torch.Size([32, 32]) torch.Size([32, 1, 32])
```

```python
x = torch.FloatTensor([[1,2,3],[4,5,6]])
y = torch.FloatTensor([[-1,-2,-3],[-4,-5,-6]])
xy1 = torch.cat([x,y], dim = 0)  # 0번째에 결합
xy2 = torch.cat([x,y], dim = 1)  # 1번째에 결합
print(x.size(), y.size(), xy1.size(), xy2.size())
```

```python
x1,x2 = torch.split(xy1,2,dim=0)
print(x1)
print(x2)
```

```python
x1,x2 = torch.split(xy1,2,dim=1)
print(x1)
print(x2)
```

### Pytorch 이미지 불러오기

- 마지막으로 torch에서 모델 학습을 위한 이미지 데이터를 불러오는 과정에 대해 학습하겠습니다.
- 이번에 이용할 데이터셋은 구글에서 제공하는 5종의 꽆 이미지가 담긴 데이터셋입니다. 이를 모델에 학습이 가능한 데이터셋으로 전처리 하는 과정을 진행하는 방법을 알아보도록 하겠습니다.

```python
import tensorflow as tf
import pathlib
dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
data_dir = tf.keras.utils.get_file(origin=dataset_url, 
                                   fname='flower_photos', 
                                   untar=True,
                                   )
data_dir = pathlib.Path(data_dir)
```

```python
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, random_split
from torchvision.transforms import transforms as T
t = T.Compose([T.Resize((224,224)),T.ToTensor()])
dataset = ImageFolder(root = data_dir,transform = t)
```

- torch에서 불러온 데이터는 numpy에서의 경우와 다른 순서를 가집니다. h w c -> c h w

```python
dataset[0][0].shape ## c,h,w로 바뀜
```

```python
train_size = int(dataset.__len__() * 0.8)
valid_size = dataset.__len__() - train_size

training_data, valid_data = random_split(dataset, [train_size, valid_size])

train_dataloader = DataLoader(training_data, batch_size=16, shuffle=True)
valid_dataloader = DataLoader(valid_data, batch_size=16, shuffle=False)
```

- 각각의 데이터를 훈련과 검증용으로 분류해 변수에 담아줍니다. 이를 데이터로더를 이용해 원하는 batch크기만큼 적재합니다.
- 데이터로더를 활용해 각각의 이미지데이터를 불러올때 batch_size에서 정의한 만큼 불러올 수 있습니다. 
- 한개의 batch_size를 불러와 시각화를 해보도록 하겠습니다.

```python
class_names = ['daisy','dandelion','roses','sunflowers','tulibs']
plt.figure(figsize=(10, 10))
images, labels = next(iter(train_dataloader))
for i in range(16):
  ax = plt.subplot(4, 4, i + 1)
  img = images[i]
  img = (img - img.min())/(img.max() - img.min())
  plt.imshow(img.permute(1, 2, 0).numpy())
  plt.title(f'Class:{class_names[labels[i]]}')
  plt.axis("off")
```
