### Numpy

numpy는 1차원 이상의 숫자 데이터를 연산하기 위해 만들어진 모듈로 벡터, 행렬 등의 연산을 가능하게 하는 모듈입니다.

numpy를 통해 만들어진 자료형을 넘파이 배열(np.array)라고 부릅니다.

```python
import numpy as np
```

np.array를 이용하여 숫자가 들어간 리스트를 넘파이 배열로 변환해 줄 수 있습니다.

list 형태에서 불가능했던 사칙연산들이 넘파이 배열 구조에서는 가능해집니다.

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
```

넘파이 배열의 연산은 elementwise로 동작합니다. 각 배열에 대해 대응되는 원소들 끼리 연산이 진행됩니다.

```python
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

또한, 1차원의 넘파이 배열은 벡터로써의 기능도 겸하며, 내적과 같은 벡터 연산이 가능합니다.

```python
np.dot(arr1,arr2) #내적
```

이번에는 넘파이 배열을 이용하여 2차원의 배열을 생성해보도록 하겠습니다.

```python
mat1 = np.array([[1,2],[3,4]])
mat2 = np.array([[5,6],[7,8]])
```

2차원 배열 역시 기본적인 연산은 elementwise하게 진행됩니다.

```python
print(mat1 + mat2)
print(mat1 - mat2)
print(mat1 * mat2)
print(mat1 / mat2)
```

2차원의 넘파이 배열은 행렬로써의 기능도 겸하며, 행렬곱과 같은 행렬 연산이 가능합니다

```python
np.matmul(mat1,mat2) ##행렬의 곱셈
```
